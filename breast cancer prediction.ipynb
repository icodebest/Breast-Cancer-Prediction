## Breast Cancer Prediction Using Machine Learning

```python
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns
% matplotlib inline
```

```python
breast=pd.read_csv("breast cancer.csv")
breast.head()
```

```python
breast.shape
```

```python
breast.duplicated().sum()
breast.info()
```

```python
breast.isnull().sum()
```

```python
breast['diagnosis'].value_counts()
```

```python
breast.drop('Unnamed: 32',axis=1,inplace=True)
breast.head()
```

```python
breast.describe()
```

## Encoding the data

```python
breast['diagnosis']=breast['diagnosis'].map({'M':1,'B':0})
breast.head(10)
```

```python
breast['diagnosis'].value_counts()
```

## Spliting data into training and testing

```python
X=breast.drop('diagnosis',axis=1)
y=breast['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=45)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
sc.fit(X_train)
```

```python
X_train=sc.transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_test
```

## Model Training

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
```

```python
y_pred
```

```python
# check the accuracy of our model with the original test data
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
